```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import pylab 
from IPython.display import display
```

```python
df = pd.read_csv('flows.txt', sep='\t', header=None, names=['packets','bytes'])

plt.figure(figsize=(10,8))
plt.plot(df.loc[:,'packets'])
plt.title('packets plot')
plt.xlabel('n')
plt.ylabel('packets')

plt.figure(figsize=(10,8))
plt.plot(df.loc[:,'bytes'])
plt.title('bytes plot')
plt.xlabel('n')
plt.ylabel('bytes')

ans = pd.DataFrame({'mean':[df['packets'].mean(),df['bytes'].mean()],
    'median':[df['packets'].median(),df['bytes'].median()]})
ans.index = ['packets', 'bytes']
display(ans)

```

```python
def running_average(data,n):
    # data: Series
    mean_n = data[0:n].mean()
    return mean_n


bytes_data = df['bytes']
mean_l = []
for n in list(range(1,len(bytes_data))):
    mean_l.append(running_average(bytes_data, n))
mean_arr=np.array(mean_l)
plt.figure(figsize=(8,6))
plt.plot(mean_arr)
plt.xlabel('x(number of passed flows)')
plt.ylabel('y(running mean in bytes)')
plt.title('running mean statistic plot')
```
