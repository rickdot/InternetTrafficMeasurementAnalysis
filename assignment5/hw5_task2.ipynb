```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import pylab 
from IPython.display import display
import random
import statsmodels.api as sm
```

```python
# original data
f = open('sampling.txt')
data = []
# samples_log = []
for line in f:
    value = float(line.split('\n')[0])
    data.append(value)
    # samples_log.append(float(np.log10(value)))

plt.figure(figsize=(10,8))
plt.hist(data, bins=100)
plt.title('histogram')
plt.xlabel('inter-arrival time')
plt.ylabel('count')

# compute the mean value
data = np.array(data)
print('Mean value of original data:')
print(np.mean(data))
```

```python
# select 5000 random
df = pd.DataFrame(data)
samples_df = df.sample(n=5000)
# display(samples_df)
plt.figure(figsize=(10,8))
plt.hist(samples_df, bins=100)
plt.title('histogram')
plt.xlabel('inter-arrival time')
plt.ylabel('count')
# compute the mean value
samples = np.array(samples_df)
print('Mean value of original data:')
print(np.mean(samples))
```

```python
# 10000 times n random value
# n = 5, 10,100
mean_v = [[],[],[]]
n = [5, 10, 100]
for j in [0,1,2]:
    ni = n[j]
    for i in range(1,10001):
        samples_df = df.sample(n=ni)
        mean_v[j].append(samples_df.iloc[:,0].mean())
    mean_v[j] = pd.Series(mean_v[j])

mean_n5 = mean_v[0]
mean_n10 = mean_v[1]
mean_n100 = mean_v[2]

# histogram
# plt.figure(figsize=(10,8))
# plt.hist(mean_n5, bins=100)
# plt.title('histogram of n=5')
# plt.xlabel('mean of n elements')
# plt.figure(figsize=(10,8))
# plt.hist(mean_n10, bins=100)
# plt.title('histogram of n=10')
# plt.xlabel('mean of n elements')
# plt.figure(figsize=(10,8))
# plt.hist(mean_n100, bins=100)
# plt.title('histogram of n=100')
# plt.xlabel('mean of n elements')

# Q-Q plot
plt.figure(figsize=(10,8))
# stats.probplot(mean_n5, dist="norm", plot=pylab)
sm.qqplot(mean_n5)
pylab.show()
plt.figure(figsize=(10,8))
# stats.probplot(mean_n10, dist="norm", plot=pylab)
sm.qqplot(mean_n10)
pylab.show()
plt.figure(figsize=(10,8))
# stats.probplot(mean_n100, dist="norm", plot=pylab)
sm.qqplot(mean_n100)
pylab.show()

# mean and standard deviation
print('mean of 10000 value(n=5, 10, 100):')
print(mean_n5.mean(), mean_n10.mean(), mean_n100.mean())
print('standard deviation of 10000 value((n=5, 10, 100)):')
print(mean_n5.std(), mean_n10.std(), mean_n100.std())


#??????????
```
