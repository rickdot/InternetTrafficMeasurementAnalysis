```python
import csv
import pandas as pd
import re
from pandas.core.frame import DataFrame
```

```python
def display_csv_data(file_path, delimiter):
    csv_data = pd.read_csv(file_path, delimiter=delimiter)
    csv_data.shape
    return csv_data

# space-separeted
display_csv_data('log_tcp_complete', ' ')
```

```python
# 1.2
def avg_calculator(file_path, column):
    data = pd.read_csv(file_path, delimiter=' ')
    avg = data.iloc[:, column].mean()
    print(avg)

cols = [3, 7, 10, 17, 21, 24]
for i in cols:
    avg_calculator('log_tcp_complete', i-1)
```

```python
# 1.3
import numpy as np
data = pd.read_csv('log_tcp_complete', delimiter=' ')
cols = [10, 7]
thresholds = [0.01, 0.1, 0.2]
record_number = len(data)
for t in thresholds:
    exceed_case = 0
    for line in range(0, len(data)):
        a = data.iloc[line, cols[0]-1]
        b = data.iloc[line, cols[1]-1]
        # ignore the case when c7 = 0
        if b == 0:
            continue
        if a/b > t:
        #if np.true_divide(a,b) > t:
            exceed_case += 1
    percentage = exceed_case/record_number
    print(percentage)
```

```python
# 1.4
def largest_value(file_path, column):
    data = pd.read_csv(file_path, delimiter=' ')
    max_value = data.iloc[:,column].max()
    return max_value

cols = [3, 9, 17, 23, 31]
for i in cols:
    max_value = largest_value('log_tcp_complete', i-1)
    print(max_value)
```
